# Question 1

Find minimum (i.e. both the point $x^∗$ and the function value $f^∗ = f(x^∗)$) of function with respect to parameters $a, b, c :$
$$f (x) = ax^2 + bx + c$$

**Solution: Q1**

Objective function: $ \quad \min_{x} f(x) $

$$ f(x) = ax^{2}+bx+c $$

First derivative,

\begin{aligned} 
f'(x) &= 2ax+b \\
2ax+b &= 0 \quad \quad \text{; since } f'(x)=0 \\
x &= \frac{-b}{2a} \\
\end{aligned} 

Second derivative, 

\begin{aligned} 
f''(x) &= 2a\\
2a &> 0 \quad \quad \text{; since } f''(x)>0 \\
a &> 0\\
\end{aligned} 

Then,
\begin{aligned}  
x^* &= \frac{-b}{2a} \quad \quad ; a>0 \\
f(x^*) &= a \left( \frac{-b}{2a} \right) ^2 + b\left( \frac{-b}{2a} \right)+c \\
&= \frac{-b^2}{4a} + c
\end{aligned}

So that,
$$ (x^*, f(x^*)) = (\frac{-b}{2a}, \frac{-b^2}{4a} + c); \quad a>0, b ∈ ℝ$$ 

Moreover, if $a=b=0, f(x) = f_{min}(x) = f_{max}(x) = c;$ $\quad c$ is constant.

# Question 2

What are dimensions of gradient of a function $h(x) = f (Ax)$, constructed of function $f : ℝ^m → ℝ$
and matrix $A ∈ ℝ^{m×k}$? 

**Solution: Q2**

Since $h(x) = f(Ax)$, using chain rule:

$$ h^{'}(x)=f^{'}(Ax) A $$

but gradient is a column vector,

$$ ∇_xh(x) = h^{'}(x)^T = A^T ∇_x f(Ax) $$

Because $\quad A^T ∈ ℝ^{k×m} \quad $ and $ f : ℝ^m → ℝ $,

so that $\quad ∇_xh(x) ∈ ℝ^{k×1} $.

# Question 3

Derive Hessian matrix $∇^2_x f(x)$ for the function $f(x) = g(Ax+b)$, assuming differentiable $g : ℝ^m → ℝ$, with dimensions $A ∈ ℝ^{m×n}; b ∈ ℝ^m; x ∈ ℝ^n$.

**Solution: Q3**

\begin{aligned}
f(x) &= g(Ax+b) \\ 
f^{'}(x) &= g^{'}(A x+b) A \quad \text{; by using chain rule} \\
∇_x f(x) &= f^{'}(x)^T = A^T ∇_x g(Ax+b) 
\end{aligned}

So that,
$$ ∇^2_x f(x) = A^T ∇_x^2 g(Ax+b) A $$

# Question 4

Solve an optimal step size problem for the quadratic function, with symmetric positive definite matrix $A ≻ 0 ∈ ℝ^{n×n}$, and $x, b, d ∈ ℝ^n$. Your goal is to find optimal $γ^∗$ for given $A, b, d, x$. The resulting expression must be written in terms of inner products (. . . , . . .) 
$$ f(γ) = (A (x + γd) , x + γd) + (b, x + γd) → \min_{γ∈ ℝ} $$

**Solution: Q4**

\begin{aligned}
f(γ) 
&= (A (x+γd) + b , x+γd) \\
&= (A (x+γd), x) + (A(x+γ d), γd) + (b,x) + (b, γd) \\
&= (Ax + Aγd, x) + (Ax + Aγd, γd) + (b,x) + (b,γd) \\
&= (Ax,x) + (Aγd,x) + (Ax,γd) + (Aγd,γd) + (b,x) + (b,γd) \\
&= (Ax,x) + γ(Ad,x) + γ(Ax,d) + γ^2(Ad,d) + (b,x) + γ(b,d) \quad \text{; since } γ ∈ ℝ
\end{aligned}

Consider, 
$$ (Ad,x) = (d,A^Tx) = (d,Ax) = (Ax,d) \quad ; A\text{ is symmetric positive definite}$$

Then,
$$ f(γ) = (Ax,x) + 2γ(Ax,d) + γ^2(Ad,d) + (b,x) + γ(b,d) $$

Find optimal $ γ^∗ $,

\begin{aligned}
f'(γ) &= 2(Ax,d) + 2γ(Ad,d) + (b,d) \\
0 &= 2(Ax,d) + 2γ(Ad,d) + (b,d) \\
γ^* &= -\frac{ 2(Ax,d) + (b,d) }{ 2(Ad,d) }
\end{aligned}

So that, 
$$ γ^* = -\frac{ (2Ax+b,d)}{ 2(Ad,d) } $$

# Question 5

Derive subgradient (subdifferential) for the function $f (x) = [x^2 − 1]_+, x ∈ ℝ$. (Do not write the subgradient method.)

**Solution: Q5**

Subgradient:
$$ \partial f(x) = \{ g :f(y) \geq f(x) + g^T(y-x), y ∈ dom(f)\} $$

<br>
Since $ f(x) = [x^2 - 1]_+ = \max\{0 , x^2 - 1\} , \quad x ∈ ℝ$.

By subgradient,
*   for $x>1$; $\quad \partial f(x)  = \{∇ (x^2 - 1)\} = 2x$
*   for $x<-1$; $\quad \partial f(x)  = \{∇ (x^2 - 1)\} = 2x$
*   for $x ∈ (-1,1)$; $\quad \partial f(x)  = \{∇ (0)\} = 0$

Consider $x=-1$ and $x=1$,
*   for $x=-1$; $\quad \partial f(-1) = \text{all slopes satisfying in range} [-2,0]$
*   for $x=1$; $\quad \partial f(-1) = \text{all slopes satisfying in range} [0,2]$

So that,

\begin{aligned}
\partial f(x) =
\begin{cases}
2x, & \text{;}\ x>1 \text{ or } x<-1\\
0, & \text{;}\ x ∈ (-1,1) \\
[-2,0], & \text{;}\ x=-1 \\
[0,2], & \text{;}\ x=1 \\ 
\end{cases}
\end{aligned}

# Question 6

Find the minimizer of $f (x, y) = x^2 + xy + 10y^2 − 22y − 5x$ numerically by steepest descent.

1. For each iteration, record the values of x, y, and f and include in a table.

**Solution: Q6**

\begin{aligned}
f(x,y) &= \frac{1}{2} 
\left[
  \begin{array}{ll}
  x & y
  \end{array} \right]
\left[
  \begin{array}{ll}
  2 & 1 \\
  1 & 20
  \end{array} \right]
\left[
  \begin{array}{l}
  x \\ y
  \end{array} \right]
-\left[ 
  \begin{array}{ll}
  5 & 22
  \end{array} \right]
\left[
  \begin{array}{l}
  x \\ y
  \end{array} \right] \\
f(x,y) &= \frac{1}{2} X^TAX - b^T X \quad ; X = \left[ \begin{array}{l} x \\ y \end{array}\right]
\end{aligned}

Calculate gradient,

\begin{aligned}
∇f(x,y) = AX - b &= 0 \\
X &= A^{-1}b = \left[ \begin{array}{l} 2 \\ 1 \end{array} \right]
\end{aligned}

Consider $ ∇^2 f(x,y) ≻ 0 $, we get $ A ≻ 0 $ (proof **), 

so that,
$$ f_{min}(x^*,y^*)= -16 $$

```python
# (proof **)
import numpy as np 
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

A = np.array([[2, 1], [1, 20]])
eig_val, _ = np.linalg.eig(A)
eig_val  # all eigen value > 0 
```

```python
A = np.array([[2, 1], [1, 20]])
b = np.array([5, 22])

def f(point):
    point = np.array(point)
    return (1/2) * point@A@point.T - b.T@point

def grad_f(point):  
    point = np.array(point)
    return A@point - b

def steepest_descent(init_point, tolerance = 1e-6, iteration = 1000):
    x_k = np.array(init_point)

    list_x = []
    list_x.append(x_k[0])

    list_y = []
    list_y.append(x_k[1])

    fx = []
    fx.append(f(x_k))

    step_size = []

    for i in range(1, iteration): 
        grad_x_k = grad_f(x_k)
        if np.linalg.norm(grad_x_k) < tolerance:   
            break

        # calculate step size
        step = (grad_x_k@grad_x_k) / (A@grad_x_k@grad_x_k)
        step_size.append(step)
        
        # update x
        x_k_1 = x_k - step*grad_x_k

        x, y = x_k_1
        list_x.append(x)
        list_y.append(y)
        fx.append(f(x_k_1))
        x_k = x_k_1
  
    df = pd.DataFrame(np.array(list_x), columns = ['x']) 
    df['y'] = np.array(list_y)
    df['f'] = np.array(fx)

    return df, i, list_x, list_y, fx, step_size
```

# Question 7

Let the cost function of the unconstrained optimization problem of interest be:
$$f(x) = \frac{1}{4} (x_1 − 1)^2 + \sum_{i=2}^{n} (2x_{i−1}^2 − x_i − 1)^2.$$
Consider two different scenarios, first $n = 3$, and then $n = 10$. Recall that the steepest descent algorithm is, 
$$ x^{k+1} = x^k − α_k∇f(x^k), α_k ∈ ℝ_{>0} $$

1. Using $x^0 = [−1.5, 1, ..., 1]^T$ write out the first iteration of the steepest descent algorithm and obtain the optimum value for $α^0$. What is the value of $x^1$ if you implement $α^0$? Verify that $f (x^1) < f (x^0)$.

```python
from scipy.optimize import minimize_scalar

def fn(x):
    first_part = (1/4)*(x[0] - 1)**2 
    second_part = 0
    for i in range(1, len(x)):
        second_part += (2*x[i-1]**2 - x[i] - 1)**2
    return first_part + second_part

def grad_fn(x): 
    grad_list = []
    for i in range(len(x)):
        if i==0:
            grad = (1/2)*(x[i]-1) + 2*(2*x[i]**2 - x[i+1] - 1)*(4*x[i])
            grad_list.append(grad)
        elif i==(len(x)-1):
            grad = 2*(2*x[i-1]**2 - x[i] - 1)*(-1)
            grad_list.append(grad)
        else:
            grad = 2*(2*x[i-1]**2 - x[i] - 1)*(-1) + 2*(2*x[i]**2 - x[i+1] - 1)*(4*x[i])
            grad_list.append(grad)
    return np.array(grad_list)

def cal_alpha(alpha, x_k, grad_x_k):
    opt = x_k - alpha*grad_x_k
    return fn(opt)

def steepest_descent_fn(x0, alpha, tolerance = 1e-5, iteration = 1000):
    x_k = np.array(x0)
    list_x = []
    list_x.append(x_k)
    fx = []
    fx.append(fn(x_k))
    fx_alpha = []

    if alpha=="optimal":
        alp_loop = True
    else:
        alp_loop = False
        alpha_opt = alpha

    for i in range(1, iteration): 
        grad_x_k = grad_fn(x_k)
 
        # calculate step size
        if alp_loop:
            alp = minimize_scalar(cal_alpha, args=(x_k, grad_x_k))
            alpha_opt = alp.x
            fx_alpha.append(alpha_opt) 
     
        fx_alpha.append(alpha_opt)

        # update x
        x_k_1 = x_k - alpha_opt*grad_x_k

        if np.linalg.norm(x_k_1 - x_k) < tolerance:
            break

        list_x.append(x_k_1)
        fx.append(fn(x_k_1))

        x_k = x_k_1

    return list_x, fx, i, fx_alpha
```
